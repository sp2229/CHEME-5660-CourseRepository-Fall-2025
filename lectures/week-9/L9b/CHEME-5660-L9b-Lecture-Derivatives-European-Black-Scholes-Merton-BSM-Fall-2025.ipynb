## What is a Derivative?
A derivative is a financial contract between a buyer and a seller that describes an agreement to transact an asset at a predetermined price on a specified future date, whose value is based on the performance of an underlying asset. 

Let's watch a short video to understand the concept of derivatives better: [Derivatives Explained in 2 Minutes in Basic English | 2023](https://www.youtube.com/watch?v=B0jeGk5GJX0).

Common types of derivatives include:
* __Futures contracts__ are legal agreements to buy or sell an underlying asset, such as commodities, equities, or other types of securities, at a predetermined price on a future date. Futures are exchange-traded.
* __Forward contracts__ are similar to futures but are traded over-the-counter (OTC) and have more flexibility in contract terms; they may have non-standard contract terms.
* __Swaps__ are derivative contracts in which one party exchanges or swaps one asset's value or cash flows for another. Swaps are traded over-the-counter (OTC). 
* __Options contracts__ give the contract buyer the right, but not the obligation, to buy (call option) or sell (put option) an asset at a specific price. Options are exchange-traded. The largest exchange for options: [Chicago Board Options Exchange (CBOE)](https://www.cboe.com)

## Call and Put Options Contracts
The two types of options contracts (derivatives whose underlying asset is typically a stock or an ETF) are: __call options__ and __put options__. Let's watch a short video to understand call and put options better: [Call versus Put Options | SoFi](https://www.youtube.com/watch?v=bvm-zeL_u8s).

What is the business case for __long and short call options__?

> __Buyer (long)__. From the buyer's perspective, a Call contract allows:
> * The buyer benefits from the upside price movement of XYZ without purchasing XYZ. Long calls have defined risk: the maximum loss for the buyer is the premium paid.
> * Long call options are an approach to purchasing XYZ at the strike price instead of the market price. 
>
> __Seller (short)__. From the seller's perspective, a Call contract offers:
> * The seller sells a call contract to collect the contract premium. Short call contracts have unlimited upside risk: call options are often only sold by investors who already own the required number of shares of XYZ as a covered call (100 shares per contract). 
> * Short call options allow the contract seller to sell shares of XYZ at a strike price of K instead of the market price S.

What is the business case for __long and short put options__?

> __Buyer (long)__. From the buyer's perspective, a Put contract allows:
> * The buyer benefits from the downside price movement of XYZ without purchasing XYZ. Long puts have defined risk: the maximum loss for the buyer is the premium paid for the contract.
> * Long put options are an approach for selling shares of XYZ at the strike price K instead of the market price S.
>
> __Seller (short)__. From the seller's perspective, a Put contract offers:
> * The seller sells a put contract to collect the contract premium. Short put contracts have unlimited downside risk: put options are often only sold by investors with funds to purchase XYZ shares (100 shares per contract). This is a cash-secured put position.
> * Short put options allow the contract seller to purchase shares of XYZ at a strike price of K instead of the market price S.

Ok, so I'm long (short) call or put options — how does the __payoff__ for these options work? There are multiple __styles__ of options contracts:
* __American-style options__ can be exercised at any time before or on the expiration date.
* __European-style options__ can only be exercised on the expiration date.
