## Curious: Where do the expressions for the up and down factors come from in the CRR model?
In the CRR model the $u$ and $d$ are based on the volatility $\sigma$ of the underlying asset, the time step $\Delta{t}$ and the symmetric condition $ud = 1$:
$$\begin{align*}
u & = e^{\sigma\sqrt{\Delta{t}}} \\
d & = e^{-\sigma\sqrt{\Delta{t}}}
\end{align*}$$
where $\sigma$ is the implied volatility of the growth rate (return) of the underlying asset. But where do these expressions come from?
___

## Derivation
The CRR model matches the volatility of the binomial model to that of geometric Brownian motion (GBM). In the binomial model, if the stock moves up by a factor $u$ or down by a factor $d$, the log-return over one step is:
$$
\ln\frac{S(t+\Delta t)}{S(t)} = 
\begin{cases}
\ln(u) = a & \text{with probability } p,\\
\ln(d) = -a & \text{with probability } 1-p,
\end{cases}
$$
where we define $a = \ln(u)$ and use the symmetry condition $d = 1/u$, which gives $\ln(d) = -a$. Let $X$ denote this log-return random variable:
$$
\begin{align*}
X & =
\begin{cases}
+a & \text{with probability } p,\\
-a & \text{with probability } 1-p.
\end{cases}
\end{align*}
$$
We want to match the variance of this two-point random variable $X$ to the variance of the GBM log-return. The mean of $X$ is:
$$
\mathbb{E}[X] = p\cdot(+a) + (1-p)\cdot(-a) = (2p-1)a.
$$
The second moment is:
$$
\mathbb{E}[X^2] = p\cdot a^2 + (1-p)\cdot a^2 = a^2.
$$
Using the variance formula $\mathrm{Var}(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2$, we have:
$$
\mathrm{Var}(X)
= \mathbb{E}[X^2] - (\mathbb{E}[X])^2
= a^2 - \big[(2p-1)a\big]^2
= a^2\Big[1 - (2p-1)^2\Big]
= 4p(1-p)a^2.
$$
From the geometric Brownian motion model, the variance of the log return over time $\Delta t$ is:
$$
\mathrm{Var}\Bigg[\ln\frac{S(t+\Delta t)}{S(t)}\Bigg]
= \sigma^2 \Delta t.
$$
We set the binomial variance equal to the GBM variance:
$$
4p(1-p)a^2 = \sigma^2 \Delta t.
$$
At this point, CRR makes a simplifying choice: at the microscopic time scale $\Delta t$, take the up/down moves to be balanced, i.e., $p = \tfrac{1}{2}$. Plugging $p = \tfrac{1}{2}$ gives:
$$
4\cdot \tfrac{1}{2}\cdot\tfrac{1}{2}a^2 = a^2 = \sigma^2 \Delta t.
$$
which yields:
$$
a^2 = \sigma^2 \Delta t
\quad\Longrightarrow\quad
a = \sigma\sqrt{\Delta t}.
$$
Recall that we defined $u = e^{a}$ and $d = e^{-a}$ from the symmetry condition $ud = 1$. Therefore:
$$
\boxed{
u = e^{+\sigma\sqrt{\Delta t}},
\qquad
d = e^{-\sigma\sqrt{\Delta t}}.
}
$$
That is exactly the Cox–Ross–Rubinstein form!
