## Concept Review: CRR Binomial Pricing Model
American style options contracts allow the holder to exercise the option __at any time__ before or on the expiration date. The standard method to price American options is the Cox, Ross, Rubinstein (CRR) binomial model.

__Reference:__
* [Cox, J. C.; Ross, S. A.; Rubinstein, M. (1979). "Option pricing: A simplified approach". Journal of Financial Economics. 7 (3): 229. CiteSeerX 10.1.1.379.7582. doi:10.1016/0304-405X(79)90015-1](https://www.sciencedirect.com/science/article/pii/0304405X79900151)

The CRR model assumes that the movement of the share price of the underlying asset can be modeled as a binomial tree, where at each time step, the price can either move up by a factor of $u$ or down by a factor of $d$. The probability of an up move is the __risk-neutral probability__ $q$, and the probability of a down move is $(1-q)$.

The __risk-neutral probability__ $q$ is calculated as:
$$
\boxed{
\begin{align*}
q = \frac{\mathcal{D}_{1,0}\left(\bar{r}\right) - d}{u - d}\quad\blacksquare
\end{align*}}
$$
where $\mathcal{D}_{1,0}\left(\bar{r}\right)$ is the discount factor for one time step at the risk-free rate $\bar{r}$, and the up and down factors $u$ and $d$ are determined based on the volatility of the underlying asset and the time step size:
$$
\boxed{
\begin{align*}
u &= e^{\sigma \sqrt{\Delta t}} \\
d &= e^{-\sigma \sqrt{\Delta t}}
\end{align*}}
$$

### Decision-Rule for American Options
The key difference between American and European options is the ability to exercise the option at any time before expiration. This introduces an additional decision-making layer in the pricing process.

>__Decision Rule__: At each node in the binomial tree, the holder of an American option must decide whether to exercise the option immediately or to hold onto it for potential future gains. The decision is based on comparing the immediate exercise value with the expected value of holding the option discounted back to the present value.

## Extrinsic versus Intrinsic Value of American Options

When evaluating American options, it's crucial to distinguish between intrinsic and extrinsic value. What do these terms mean in the context of American options?

The premium (price) of an American option can be broken down into two components: intrinsic value and extrinsic value.

* __Intrinsic Value__: The intrinsic value of an American option is the __immediate exercise value__. For a call option, it's calculated as $\max(0, S - K)$, where $S$ is the current stock price and $K$ is the strike price. For a put option, it's $\max(0, K - S)$.
* __Extrinsic Value__: The extrinsic value, also known as time value, represents the additional value of the option based on the potential for future price movements before expiration. It is calculated as the difference between the total option price and its intrinsic value.

The extrinsic value $E_{V}(t)$ can be expressed mathematically as:
$$
\boxed{
\begin{align*}
E_{V}(t) = \text{Option Price} - \text{Intrinsic Value}\quad\blacksquare
\end{align*}}
$$
or in expanded form:
$$
\boxed{
\begin{align*}
E_{V}(t) &= \begin{cases}
\mathcal{P}_{c}(K,S(0)) - \max(0, S(t) - K) & \text{(for Call Options)} \\
\mathcal{P}_{p}(K,S(0)) - \max(0, K - S(t)) & \text{(for Put Options)}
\end{cases}
\end{align*}}
$$
Ok, so this has a strange sort of edge case: when the option is __out of the money__, i.e., the options has no intrinsic value, the entire option price is extrinsic value. Why would we be willing to pay for an option that has no intrinsic value? 

> __Why do people pay money for something that is "worthless"?__ The answer lies in the potential for future price movements. Even if an option is currently out of the money, there is still a chance that the underlying asset's price could move favorably before expiration, making the option valuable.

As time moves towards expiration, the extrinsic value of an option tends to decrease, a phenomenon known as "time decay." This is because there is less time for the underlying asset's price to move in a favorable direction.
