## Motivation: Are Treasury securities really risk-free?
United States Treasury securities are considered risk-free, backed by the country's full faith and credit. When the U.S. government borrows, it commits to repaying under specific terms, which it has consistently honored. In contrast, this guarantee does not apply to Treasury securities from other countries.

Are Treasury securities risk-free? No, there are two types of risk:
* __Default risk__: In this scenario, the issuer of the bill/note/bond does not pay back the debt, i.e., they default. While the United States has never defaulted, many examples of countries in modern times have defaulted. For example, [Greece defaulted in 2011](https://en.wikipedia.org/wiki/Greek_government-debt_crisis) and [Russia defaulted in 1998](https://en.wikipedia.org/wiki/1998_Russian_financial_crisis). This makes borrowing more expensive (if not impossible) because the issuer has _poor credit_. For U.S. treasury securities, the default risk is __low__.
* __Interest rate risk__: In this scenario, we (the debt holder) want to sell the security on the secondary market, i.e., to convert the debt security back into cash. However, because market conditions, i.e., interest rates, have changed since the debt was issued (or we purchased it), the note/bill/bond price may have changed. Thus, we could experience a realized capital loss (gain) when we sell the bill/note/bond. For U.S. treasury securities, the interest rate risk is __high__.

Let's look at another recent case study that demonstrates _interest rate risk_.


### Case Study: Interest Rate Risk and Silicon Valley Bank (SVB)
Silicon Valley Bank collapsed in March 2023 after rising interest rates eroded the value of its long‐duration bond portfolio. When a sudden run on deposits by tech clients occurred, the rapid outflow of funds left the bank unable to meet liquidity needs, leading to its closure by regulators.
* __What happened__? Silicon Valley Bank (SVB) invested heavily in long-duration U.S. government securities while interest rates were low. As the Federal Reserve raised rates in 2022–2023, the value of these bonds fell, leading to significant unrealized losses. In early March 2023, many uninsured tech-startup depositors withdrew funds. When SVB announced a capital raise to cover its realized bond losses, panic ensued, resulting in bank failure. 
* __What was the aftermath?__ The [Federal Deposit Insurance Corporation (FDIC)](https://www.fdic.gov) took control of SVB on March 10, 2023, marking the most significant bank failure since the 2008 financial crisis (second largest in U.S. history). To prevent a broader banking crisis, the U.S. government stepped in to protect all depositors, including those with uninsured balances. The [United States Federal Reserve](https://www.federalreserve.gov) also introduced emergency lending facilities to provide liquidity to banks facing similar issues.

## Zero-Coupon Treasury Bill

Let's start with the simplest case—a Treasury bill that makes no interest payments, just returns your money plus profit at maturity.

> **Why start here?** T-bills are like the "hydrogen atom" of finance—simple enough to understand completely, but containing all the fundamental principles that apply to more complex bonds.

Previously, we showed that the price of a zero-coupon Treasury bill is given by: $V_{B} = \mathcal{D}_{nT,0}^{-1}(\bar{r})\;V_{P}$, where $V_{P}$ is the face (par) value of the bill, and $\mathcal{D}^{-1}_{nT,0}(\bar{r})$ is the (inverse) discrete multiperiod discount factor for compounding events $0\rightarrow nT$, which is given by: $\mathcal{D}^{-1}_{nT,0}(\bar{r}) = \left(1+\bar{r}/n\right)^{-nT}$.

> **Intuition**: You pay less than $1,000 today to get exactly $1{,}000 back later. The "discount factor" tells you exactly how much less to pay based on current interest rates and time to maturity.

where $n$ is the number of compounding periods per year, and $T$ is the time to maturity in years.

### Deriving the Price Sensitivity

Now comes the key question: **How does the T-bill price change when market conditions change?**

The change in the price of the bill with respect to the discount rate $\bar{r}$ and time to maturity $T$ is given by:

$$\begin{aligned} dV_{B} &= \frac{\partial V_{B}}{\partial \bar r}\,d\bar{r}+\underbrace{\frac{\partial V_{B}}{\partial \bar c}\,d\bar c}_{=\,0} +\frac{\partial V_{B}}{\partial T}\,dT\\ &= -\underbrace{V_{P}\left(1+\bar{r}/n\right)^{-nT}\left(\frac{T}{1+\bar{r}/n}\right)}_{V_{B,\bar r}}\;d\bar{r} + 0 - \underbrace{V_{P}\left(1+\bar{r}/n\right)^{-nT}\left(n\ln(1+\bar{r}/n)\right)}_{V_{B,T}}\;dT\\ &= -V_{B,\bar r}\;d\bar{r} - V_{B,T}\;dT\quad\blacksquare \end{aligned}$$

> **Key insight**: The coupon term disappears (equals zero) because T-bills don't pay coupons—they're pure discount instruments.

where $V_{B,\bar r}$ is the sensitivity of the bond price to changes in the discount rate, and $V_{B,T}$ is the sensitivity of the bond price to changes in time to maturity. Both of these sensitivities are non-negative:

$$\begin{aligned} V_{B,\bar r} &= \underbrace{V_{P}\Bigl(1+\tfrac{\bar r}{n}\Bigr)^{-nT}\;\left(\frac{T}{1+\tfrac{\bar r}{n}}\right)}_{\geq\;0}\\ V_{B,T} &= \underbrace{V_{P}\Bigl(1+\tfrac{\bar r}{n}\Bigr)^{-nT}\;\left(n\,\ln\;\Bigl(1+\tfrac{\bar r}{n}\Bigr)\right)}_{\geq\;0} \end{aligned}$$

### What This Math Actually Means

The mathematical expressions above tell us precisely how T-bill prices respond to market changes, but let's translate this into intuitive understanding.

From these expressions, we can see the direction of the price changes with respect to changes in the discount rate and time to maturity. Let's take a closer look at how the price of the bill changes with respect to these two variables.

**Inversely proportional to the discount rate**: As the discount rate increases, the price of the bill decreases, and vice versa. This is because a higher discount rate means that future cash flows are worth less in present value terms. So $V_{B}\downarrow$ as $\bar{r}\uparrow$, assuming $dT\approx 0$.

> **Example**: You buy a 52-week T-bill at auction. If the discount rate increases from 2% to 3%, the price in the secondary market will be **lower** than your purchase price, because the future cash flow (the face value) is discounted more heavily.

**Inversely proportional to time to maturity**: As the time to maturity increases, the price of the bill decreases for the same discount rate, and vice versa. Thus, $V_{B}\downarrow$ as $T\uparrow$, assuming $d\bar{r}\approx 0$. However, **in real time** as you hold the bill, calendar time $t$ advances and the remaining time $T$ **decreases** ($dT<0$):

> **Example**: You buy a 52-week T-bill at auction. Six months later ($T$ falls from 1 yr to 0.5 yr) and with $\bar r$ unchanged, the price in the secondary market will be **higher** than your purchase price—exactly because there are fewer remaining periods to discount.

## Coupon-Bearing Treasury Note or Bond

Previously, we showed that the price of a coupon-bearing Treasury note (bond) is given by:

$$V_{B} = \mathcal{D}^{-1}_{nT,0}(\bar{r})\cdot V_{P} + C\cdot\sum_{j=1}^{nT}\mathcal{D}^{-1}_{j,0}(\bar{r})$$

where $C\equiv(\bar{c}/n)\cdot V_{P}$ is the value of a coupon payment, $\bar{c}$ is the coupon rate, and $\mathcal{D}^{-1}_{j,0}(\bar{r}) = \left(1+\bar{r}/n\right)^{-j}$ is the discrete multiperiod discount factor for periods $0\rightarrow j$. The change in the price of the note (bond) with respect to the discount rate $\bar{r}$, coupon rate $\bar{c}$, and time to maturity $T$ is given by:

$$
\begin{align*}
    dV_{B} &= \frac{\partial V_{B}}{\partial \bar r}\,d\bar{r} + \frac{\partial V_{B}}{\partial \bar c}\,d\bar c + \frac{\partial V_{B}}{\partial T}\,dT\\
    &= -\underbrace{V_{P}(1+\bar{r}/n)^{-nT}\left(\frac{nT}{1+\bar{r}/n}\right)\left(\frac{1}{n}\right) + C\sum_{j=1}^{nT}(1+\bar{r}/n)^{-j}\left(\frac{j}{1+\bar{r}/n}\right)\left(\frac{1}{n}\right)}_{V_{B,\bar{r}}}\;d\bar{r}\\
    &\quad + \underbrace{\frac{V_{P}}{n}\sum_{j=1}^{nT}(1+\bar{r}/n)^{-j}}_{V_{B,\bar{c}}}\;d\bar{c}\\
    &\quad - \underbrace{V_{P}(1+\bar{r}/n)^{-nT}n\ln(1+\bar{r}/n) + C(1+\bar{r}/n)^{-nT}n\ln(1+\bar{r}/n)}_{V_{B,T}}\;dT\\
    &= -V_{B,\bar{r}}\;d\bar{r} + V_{B,\bar{c}}\;d\bar{c} - V_{B,T}\;dT\quad\blacksquare
\end{align*}
$$

where $V_{B,\bar{r}}$ is the sensitivity of the bond price to changes in the discount rate, $V_{B,\bar{c}}$ is the sensitivity of the bond price to changes in the coupon rate, and $V_{B,T}$ is the sensitivity of the bond price to changes in time to maturity. All of these sensitivities are non-negative:

$$
\begin{align*}
V_{B,\bar r}
& = \underbrace{V_{P}\Bigl(1+\tfrac{\bar r}{n}\Bigr)^{-nT}\;\left(\frac{T}{1+\tfrac{\bar r}{n}}\right) + C\sum_{j=1}^{nT}\Bigl(1+\tfrac{\bar r}{n}\Bigr)^{-j}\;\left(\frac{j}{n(1+\tfrac{\bar r}{n})}\right)}_{\geq\;0}\\
V_{B,\bar c}
& = \underbrace{\frac{V_{P}}{n}\sum_{j=1}^{nT}\Bigl(1+\tfrac{\bar r}{n}\Bigr)^{-j}}_{\geq\;0}\\
V_{B,T}
& = \underbrace{V_{P}\Bigl(1+\tfrac{\bar r}{n}\Bigr)^{-nT}\;\left(n\ln\;\Bigl(1+\tfrac{\bar r}{n}\Bigr)\right) + C\Bigl(1+\tfrac{\bar r}{n}\Bigr)^{-nT}\;\left(n\ln\;\Bigl(1+\tfrac{\bar r}{n}\Bigr)\right)}_{\geq\;0}
\end{align*}
$$
