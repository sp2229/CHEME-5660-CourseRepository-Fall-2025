## Concept Review: Are Treasury Securities Really Risk-Free?
U.S. Treasury securities are considered risk-free, backed by the country's full faith and credit. When the U.S. government borrows, it commits to repaying under specific terms, which it has consistently honored. In contrast, this guarantee does not apply to Treasury securities from other countries.

Are Treasury securities risk-free? No, there are two types of risk:
* __Default risk__: In this scenario, the issuer of the bill/note/bond does not pay back the debt, i.e., they default. While the United States has never defaulted, many countries in modern times have defaulted. For example, [Greece defaulted in 2011](https://en.wikipedia.org/wiki/Greek_government-debt_crisis) and [Russia defaulted in 1998](https://en.wikipedia.org/wiki/1998_Russian_financial_crisis). This makes borrowing more expensive (if not impossible) because the issuer has _poor credit_. For U.S. Treasury securities, the default risk is __low__.
* __Interest rate risk__: In this scenario, we (the debt holder) want to sell the security on the secondary market, i.e., to convert the debt security back into cash. However, because market conditions, i.e., interest rates, have changed since the debt was issued (or we purchased it), the note/bill/bond price may have changed. Thus, we could experience a realized capital loss (gain) when we sell the bill/note/bond. For U.S. Treasury securities, the interest rate risk is __high__.

Let's look at another recent case study that demonstrates _interest rate risk_.


### Case Study: Interest Rate Risk and Silicon Valley Bank (SVB)
Silicon Valley Bank collapsed in March 2023 after rising interest rates eroded the value of its long‐duration bond portfolio. When a sudden run on deposits by tech clients occurred, the rapid outflow of funds left the bank unable to meet liquidity needs, leading to its closure by regulators.
* __What happened__? Silicon Valley Bank (SVB) invested heavily in long-duration U.S. government securities while interest rates were low. As the Federal Reserve raised rates in 2022–2023, the value of these bonds fell, leading to significant unrealized losses. In early March 2023, many uninsured tech-startup depositors withdrew funds. When SVB announced a capital raise to cover its realized bond losses, panic ensued, resulting in bank failure. 
* __What was the aftermath?__ The [Federal Deposit Insurance Corporation (FDIC)](https://www.fdic.gov) took control of SVB on March 10, 2023, marking the most significant bank failure since the 2008 financial crisis (second largest in U.S. history). To prevent a broader banking crisis, the U.S. government stepped in to protect all depositors, including those with uninsured balances. The [U.S. Federal Reserve](https://www.federalreserve.gov) also introduced emergency lending facilities to provide liquidity to banks facing similar issues.

Let's watch a video describing the [collapse of Silicon Valley Bank from PBS](https://www.youtube.com/watch?v=7tNt7WaP4iA).

__Bond Pricing and Interest Rate Sensitivity__: SVB's collapse illustrates the fundamental relationship between bond prices and interest rates - when rates rise, bond prices fall, and the magnitude of this price change depends on the bond's characteristics. 
> __Model__: The mathematical relationships governing this sensitivity were formalized by Burton Malkiel in his famous bond pricing theorems. These theorems explain why SVB's long-duration bonds experienced such dramatic price declines and provide quantitative tools for measuring and managing interest rate risk. Understanding these relationships is crucial for any institution holding fixed-income securities.

## Zero-Coupon Treasury Bill

Let's start with the simplest case—a Treasury bill that makes no interest payments, just returns your money plus profit at maturity.

> **Why start here?** T-bills are like the "hydrogen atom" of finance—simple enough to understand completely, but containing all the fundamental principles that apply to more complex bonds.

Previously, we showed that the price of a zero-coupon Treasury bill is given by: $V_{B} = \mathcal{D}_{nT,0}^{-1}(\bar{r})\;V_{P}$, where $V_{P}$ is the face (par) value of the bill, and $\mathcal{D}^{-1}_{nT,0}(\bar{r})$ is the (inverse) discrete multiperiod discount factor for compounding events $0\rightarrow nT$. This is given by: $\mathcal{D}^{-1}_{nT,0}(\bar{r}) = \left(1+\bar{r}/n\right)^{-nT}$, where $n$ is the number of compounding periods per year, and $T$ is the time to maturity in years.

> **Intuition**: You pay less than 100 USD today to get exactly 100 USD back later. The inverse of the __discount factor__ tells you exactly how much less we pay based on current interest rates and time to maturity.

### Deriving the Price Sensitivity
Now comes the key question: **How does the T-bill price change when market conditions change?**

The change in the price of the bill with respect to the discount rate $\bar{r}$ and time to maturity $T$ is given by:
$$
\begin{aligned} 
dV_{B} &= \frac{\partial V_{B}}{\partial \bar r}\,d\bar{r}+\underbrace{\frac{\partial V_{B}}{\partial \bar c}\,d\bar c}_{=\,0} +\frac{\partial V_{B}}{\partial T}\,dT\\ &= -\underbrace{V_{P}\left(1+\bar{r}/n\right)^{-nT}\left(\frac{T}{1+\bar{r}/n}\right)}_{V_{B,\bar r}}\;d\bar{r} + 0 - \underbrace{V_{P}\left(1+\bar{r}/n\right)^{-nT}\left(n\ln(1+\bar{r}/n)\right)}_{V_{B,T}}\;dT\\ &= -V_{B,\bar r}\;d\bar{r} - V_{B,T}\;dT\quad\blacksquare 
\end{aligned}
$$

> **Key insight**: The coupon term disappears (equals zero) because T-bills don't pay coupons, they're pure discount instruments.


The term $V_{B,\bar r}$ is the sensitivity of the bond price to changes in the discount rate, and $V_{B,T}$ is the sensitivity of the bond price to changes in time to maturity. Both of these sensitivities are non-negative:
$$
\begin{aligned} 
V_{B,\bar r} &= \underbrace{V_{P}\Bigl(1+\bar{r}/{n}\Bigr)^{-nT}\;\left(\frac{T}{1+\bar{r}/{n}}\right)}_{\geq\;0}\\ 
V_{B,T} &= \underbrace{V_{P}\Bigl(1+\bar{r}/{n}\Bigr)^{-nT}\;\left(n\,\ln\;\Bigl(1+\bar{r}/{n}\Bigr)\right)}_{\geq\;0} 
\end{aligned}
$$

### What This Says to Us

The mathematical expressions above tell us precisely how T-bill prices respond to market changes, but let's translate this into intuitive understanding.

**Inversely proportional to the discount rate**: As the discount rate increases, the price of the bill decreases, and vice versa. This is because a higher discount rate means that future cash flows are worth less in present value terms. So $V_{B}\downarrow$ as $\bar{r}\uparrow$, assuming $dT\approx 0$.

> **Example**: You buy a 52-week T-bill at auction. If the discount rate increases from 2% to 3%, the price in the secondary market will be **lower** than your purchase price, because the future cash flow (the face value) is discounted more heavily.

**Inversely proportional to time to maturity**: As the time to maturity increases, the price of the bill decreases for the same discount rate. Thus, $V_{B}\downarrow$ as $T\uparrow$, assuming $d\bar{r}\approx 0$. However, **in real time** as you hold the bill, calendar time $t$ advances and the remaining time $T$ **decreases** ($dT<0$):

> **Example**: You buy a 52-week T-bill at auction. Six months later ($T$ falls from 1 yr to 0.5 yr) and with $\bar r$ unchanged, the price in the secondary market will be **higher** than your purchase price, because there are fewer remaining periods to discount.

## Coupon-Bearing Treasury Notes and Bonds

Previously, we showed that the price of a coupon-bearing Treasury note (bond) is given by:

$$V_{B} = \mathcal{D}^{-1}_{nT,0}(\bar{r})\cdot V_{P} + C\cdot\sum_{j=1}^{nT}\mathcal{D}^{-1}_{j,0}(\bar{r})$$

where $C\equiv(\bar{c}/n)\cdot V_{P}$ is the value of a coupon payment, $\bar{c}$ is the coupon rate, and $\mathcal{D}^{-1}_{j,0}(\bar{r}) = \left(1+\bar{r}/n\right)^{-j}$ is the discrete multiperiod discount factor for periods $0\rightarrow j$. The change in the price of the note (bond) with respect to the discount rate $\bar{r}$, coupon rate $\bar{c}$, and time to maturity $T$ is given by:

$$
\begin{align*}
    dV_{B} &= \frac{\partial V_{B}}{\partial \bar r}\,d\bar{r} + \frac{\partial V_{B}}{\partial \bar c}\,d\bar{c} + \frac{\partial V_{B}}{\partial T}\,dT\\
    &= -\underbrace{V_{P}(1+\bar{r}/n)^{-nT}\left(\frac{nT}{1+\bar{r}/n}\right)\left(\frac{1}{n}\right) + C\sum_{j=1}^{nT}(1+\bar{r}/n)^{-j}\left(\frac{j}{1+\bar{r}/n}\right)\left(\frac{1}{n}\right)}_{V_{B,\bar{r}}}\;d\bar{r}\\
    &\quad + \underbrace{\frac{V_{P}}{n}\sum_{j=1}^{nT}(1+\bar{r}/n)^{-j}}_{V_{B,\bar{c}}}\;d\bar{c}\\
    &\quad - \underbrace{V_{P}(1+\bar{r}/n)^{-nT}n\ln(1+\bar{r}/n) + C(1+\bar{r}/n)^{-nT}n\ln(1+\bar{r}/n)}_{V_{B,T}}\;dT\\
    &= -V_{B,\bar{r}}\;d\bar{r} + V_{B,\bar{c}}\;d\bar{c} - V_{B,T}\;dT\quad\blacksquare
\end{align*}
$$

where $V_{B,\bar{r}}$ is the sensitivity of the bond price to changes in the discount rate, $V_{B,\bar{c}}$ is the sensitivity of the bond price to changes in the coupon rate, and $V_{B,T}$ is the sensitivity of the bond price to changes in time to maturity. All of these sensitivities are non-negative:

$$
\begin{align*}
V_{B,\bar r}
& = \underbrace{V_{P}\Bigl(1+\bar{r}/n\Bigr)^{-nT}\;\left(\frac{T}{1+\bar{r}/n}\right) + C\sum_{j=1}^{nT}\Bigl(1+\bar{r}/n\Bigr)^{-j}\;\left(\frac{j}{n(1+\bar{r}/n)}\right)}_{\geq\;0}\\
V_{B,\bar c}
& = \underbrace{\frac{V_{P}}{n}\sum_{j=1}^{nT}\Bigl(1+\bar{r}/n\Bigr)^{-j}}_{\geq\;0}\\
V_{B,T}
& = \underbrace{V_{P}\Bigl(1+\bar{r}/n\Bigr)^{-nT}\;\left(n\ln\;\Bigl(1+\bar{r}/n\Bigr)\right) + C\Bigl(1+\bar{r}/n\Bigr)^{-nT}\;\left(n\ln\;\Bigl(1+\bar{r}/n\Bigr)\right)}_{\geq\;0}
\end{align*}
$$

### What This Says to Us

The mathematical expressions above tell us how coupon-bearing note (bond) prices respond to market changes, but let's translate this into intuitive understanding. Notice how the coupon payments create additional complexity compared to the simple T-bill case.

**Inversely proportional to the discount rate**: As the discount rate increases, the price of the bond decreases, and vice versa. This is the same relationship as T-bills, but now __both__ the final principal payment __and__ all the intermediate coupon payments are discounted more heavily. So $V_{B}\downarrow$ as $\bar{r}\uparrow$, assuming $d\bar{c}\approx 0$ and $dT\approx 0$.

> **Example**: You buy a 10-year Treasury note paying 3% coupons when market rates are 3%. If market rates rise to 4%, your bond's price will fall because new investors can get 4% coupons elsewhere. Both your future 3% coupon payments and the final principal are now worth less in present value terms.

**Directly proportional to the coupon rate**: As the coupon rate increases, the price of the bond increases, and vice versa. This makes intuitive sense, higher coupon payments mean more cash flow to the bondholder. Thus, $V_{B}\uparrow$ as $\bar{c}\uparrow$, assuming $d\bar{r}\approx 0$ and $dT\approx 0$.

> **Example**: Compare two 10-year Treasury notes: one paying 2% coupons and another paying 4% coupons. The 4% coupon bond will trade at a higher price because it provides twice the annual income stream, even though both have the same final par payment.

**Inversely proportional to time to maturity**: As the time to maturity increases, the price of the bond decreases for the same discount and coupon rates, and vice versa. This relationship is more complex than for T-bills because longer maturity means __more__ coupon payments (which increases value) but also means __longer__ discounting periods (which decreases value). The discounting effect dominates. Thus, $V_{B}\downarrow$ as $T\uparrow$, assuming $d\bar{r}\approx 0$ and $d\bar{c}\approx 0$. However, **in real time** as you hold the bond, calendar time $t$ advances and the remaining time $T$ **decreases** ($dT<0$):

> **Example**: You buy a 30-year Treasury bond paying 4% coupons when market rates are 5% (so you buy it at a discount to par). Five years later ($T$ falls from 30 yr to 25 yr) and with $\bar{r}$ and $\bar{c}$ unchanged, the bond's price will be **higher** than your purchase price. Why? You're now 5 years closer to all those future coupon payments and the final principal payment, so the present value of those cash flows increases.

### Connection to Malkiel's Bond Pricing Theorems

These mathematical relationships directly embody Burton Malkiel's famous bond pricing theorems:

* __Theorem 1__ (Bond prices move inversely to bond yields): This is captured by the negative sign on $V_{B,\bar{r}}$ in our total differential.
* __Theorem 2__ (Longer maturity bonds are more sensitive to yield changes): The $V_{B,\bar{r}}$ expression shows that sensitivity increases with $T$ because longer bonds have more periods affected by the rate change.
* __Theorem 4__ (Price movements are asymmetric): The nonlinear discount factors in our expressions mean that rate decreases cause larger price increases than rate increases cause price decreases.
* __Theorem 5__ (Higher coupon bonds are less sensitive to yield changes): Bonds with higher $\bar{c}$ have more of their value concentrated in earlier coupon payments, making them less sensitive to distant cash flows affected by rate changes.

Our mathematical framework provides the quantitative foundation for understanding and measuring these relationships that Malkiel identified empirically.
