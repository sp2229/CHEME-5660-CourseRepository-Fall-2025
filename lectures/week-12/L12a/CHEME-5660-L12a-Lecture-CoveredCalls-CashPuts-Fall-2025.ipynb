## Covered Calls
A covered call is an options strategy where an investor holds a long position in an asset and sells call options on that same asset to generate income from the option premiums. This strategy is typically used when the investor expects the underlying asset's price to remain relatively stable or to increase slightly.

Let's watch a short video explaining the basics of covered calls from SMB Capital: [Simple & Low Risk Investing Strategy](https://www.yout-ube.com/watch?v=Tid1rfVIc9c).

> __Naked calls:__ Of course, you can benefit from selling call options without owning the underlying asset, but this is known as a __naked call__ which carries undefined downside risk. In contrast, a covered call limits potential losses because the investor already owns the underlying asset at some cost basis.

### Analysis of Covered Call Portfolios
Let's look at a general FEPI style covered call portfolio containing $|\mathcal{C}|$ legs of American-style __short call__ contracts on an underlying asset, all contracts have the same expiration date $T$ (which we assume is short enough to ignore time-value of money issues). The portfolio value consists of three terms: a cash position, a stock position, and an options position.
* __Cash Position:__ The cash position is simply the amount of cash held in the portfolio, denoted as $N_C$. This could be positive (cash held) or negative (cash borrowed). For now, we'll only consider $N_{C}\geq{0}$ (no borrowing).
* __Stock Position:__ The stock position is the value of the shares of the underlying stock $i$ held in the portfolio, denoted as $N^{(i)}_S S^{(i)}(t)$, where $N^{(i)}_S$ is the number of shares held of stock $i$ and $S^{(i)}(t)$ is the current share price of stock $i$ at time $t$.
* __Options Position:__ The options position is the value of the options contracts held in the portfolio, denoted as $m \sum_{i=1}^{|\mathcal{C}|} \theta_i n_i \mathcal{P}_i(S,t)$, where $m$ is the contract multiplier (shares per contract), $\theta_i$ indicates long ($1$) or short ($-1$) position, $n_i$ is the number of contracts, and $\mathcal{P}_i(S,t)$ is the price of the option leg $i$ at time $t$.

Thus, the general portfolio value at time $t$ is given by:
$$
\begin{align*}
\Pi(S^{(1)},S^{(2)},\ldots,S^{(|\mathcal{C}|)},t) &= \underbrace{N_C}_{\text{Cash}} + \underbrace{\sum_{i=1}^{|\mathcal{C}|} N^{(i)}_S S^{(i)}(t)}_{\text{Stock Position}} + \underbrace{m\sum_{i=1}^{|\mathcal{C}|} \theta_i n_i \mathcal{P}_i(S^{(i)}(t),K_{i})}_{\text{Options Position}}.
\end{align*}
$$
where we use $S^{(i)}(t)$ to denote the price of stock $i$ at time $t$, and $\mathcal{P}_i(S^{(i)}(t),K_{i})$ to denote the price of option leg $i$ at time $t$ with strike price $K_i$. We have not explicitly written the dependence of the option price on time $t$ for clarity.

We can also write a profit/loss (P/L) expression for the portfolio at time $t$ by subtracting the initial portfolio value $\Pi(S,0)$ from the portfolio value at time $t$ (for clarity, we'll use $S$ as shorthand for the vector of stock prices):
$$
\begin{align*}
\hat{P}(S,t) &= \Pi(S,t) - \Pi(S,0)\\
&= \left[N^{(t)}_C + \sum_{i=1}^{|\mathcal{C}|} N^{(i)}_S S^{(i)}(t) + m\sum_{i=1}^{|\mathcal{C}|} \theta_i n_i \mathcal{P}_i(S^{(i)}(t),K_{i})\right] - \left[N^{(0)}_C + \sum_{i=1}^{|\mathcal{C}|} N^{(i)}_S S^{(i)}(0) + m\sum_{i=1}^{|\mathcal{C}|} \theta_i n_i \mathcal{P}_i(S^{(i)}(0),K_{i})\right]\\
&= \underbrace{\left(N^{(t)}_C - N^{(0)}_C\right)}_{\text{Change in Cash}} + \underbrace{\sum_{i=1}^{|\mathcal{C}|} N^{(i)}_S \left(S^{(i)}(t) - S^{(i)}(0)\right)}_{\text{Change in Stock Value}} + m\underbrace{\sum_{i=1}^{|\mathcal{C}|} \theta_i n_i \left(\mathcal{P}_i(S^{(i)}(t),K_{i}) - \mathcal{P}_i(S^{(i)}(0),K_{i})\right)}_{\text{Change in Option Value}}.
\end{align*}
$$
At expiration time $t = T$, we make an assumption that __extrinsic value__ of the options contracts is zero, meaning that the options are either exercised or expire worthless. Thus, the portfolio P/L at expiration simplifies to:
$$
\begin{align*}
\hat{P}(S,T) &= \left(N^{(T)}_C - N^{(0)}_C\right) + \sum_{i=1}^{|\mathcal{C}|} N^{(i)}_S \left(S^{(i)}(T) - S^{(i)}(0)\right) + m\sum_{i=1}^{|\mathcal{C}|} \theta_i n_i \left(V^{c}_i(S,T) - \mathcal{P}_i(S^{(i)}(0),K_{i})\right)\\
&= \left(N^{(T)}_C - N^{(0)}_C\right) + \sum_{i=1}^{|\mathcal{C}|} N^{(i)}_S \left(S^{(i)}(T) - S^{(i)}(0)\right) + m\sum_{i=1}^{|\mathcal{C}|} \theta_i n_i \left(\left(S^{(i)}(T) - K_i\right)^{+} - \mathcal{P}_i(S^{(i)}(0),K_{i})\right)\\
&= \underbrace{\left(N^{(T)}_C - N^{(0)}_C\right)}_{\text{Change in Cash}} + \underbrace{\sum_{i=1}^{|\mathcal{C}|} N^{(i)}_S \left(S^{(i)}(T) - S^{(i)}(0)\right)}_{\text{Change in Stock Value}} - m\underbrace{\sum_{i=1}^{|\mathcal{C}|} n_i \left(\left(S^{(i)}(T) - K_i\right)^{+} - \mathcal{P}_i(S^{(i)}(0),K_{i})\right)}_{\text{Change in Option Value}}.
\end{align*}
$$
where $V^{c}_i(S,T) = \left(S^{(i)}(T) - K_i\right)^{+}$ is the intrinsic value of the call option at expiration, and $K_i$ is the strike price of option leg $i$, $N_{S}^{(i)} \geq m n_i$ and $\theta_{i} = -1$ for a covered call position. This expression captures the profit or loss of a covered call portfolio at expiration, taking into account changes in cash, stock value, and option value.

### Single covered call
The general expression is handy if we want to analyze a portfolio with multiple covered call legs, but let's simplify to a single covered call position to better understand the strategy. Consider a portfolio with one covered call position on stock `XYZ`. Further suppose we own exactly $N_{s} = m n$ shares of stock `XYZ` and have sold $n$ call option contracts with strike price $K$ and expiration date $T$. The portfolio profit at expiration time $t = T$ is given by:
$$
\begin{align*}
\hat{P}(S,T) &= \left(N^{(T)}_C - N^{(0)}_C\right) + N_S \left(S(T) - S(0)\right) - m n \left(\left(S(T) - K\right)^{+} - \mathcal{P}(S(0),K)\right)\\
&= \left(N^{(T)}_C - N^{(0)}_C\right) + m n \left(S(T) - S(0)\right) - m n \left(\left(S(T) - K\right)^{+} - \mathcal{P}(S(0),K)\right)\\
\end{align*}
$$
which simplifies to:
$$
\boxed{
\begin{align*}
\hat{P}(S,T) &= \begin{cases}
\left(N^{(T)}_C - N^{(0)}_C\right) + m n\;\overbrace{\left(K - S^{\dagger} + \mathcal{P}(S(0),K)\right)}^{\text{realized gain}}, & \text{if } S(T) > K\;\text{(call exercised)}\\
\left(N^{(T)}_C - N^{(0)}_C\right) + m n\;\mathcal{P}(S(0),K) + m n\;\underbrace{\left(S(T) - S^{\dagger}\right)}_{\text{unrealized gain}}, & \text{if } S(T) \leq K\;\text{(call expires worthless)}
\end{cases}
\end{align*}}
$$
where $S^{\dagger}$ is the share acquisition price. In many cases, $S^{\dagger} = S(0)$. However, when would it be different?

> __What??__
>
> If we bought the stock and sold the call option at time $t=0$, then $S^{\dagger}$ would be the stock price at time $t=0$, i.e., $S^{\dagger} = S(0)$. However, suppose we acquired the stock at an earlier time, then $S^{\dagger}$ would be the price at which we __originally purchased__ the stock, in which case $S^{\dagger} \neq S(0)$. We refer to $S^{\dagger}$ as the share acquisition price to account for both scenarios. Given my background, I tend to think of $S^{\dagger}$ as the __cost basis__ of the stock position.

What is the cash change term $(N^{(T)}_C - N^{(0)}_C)$ doing in the profit expression?

> __Note on Cash Change Term:__
>
> The term $(N_C^{(T)} - N_C^{(0)})$ captures any extraneous cash flows not directly related to the covered call strategy itself, such as deposits, withdrawals, dividends received, or interest earned on cash balances. For a pure covered call analysis focused solely on the option strategy, this term would typically be zero. 

This piecewise profit function shows two interesting regimes:

> * __Regime 1:__ In this case, $S(T) > K$, the stock price at expiration exceeds the strike price of the call option. The call option is exercised, and we must sell the shares at the strike price $K$. The profit consists of two terms (excluding cash changes): the difference between the strike price $K$ and the share acquisition price $S^{\dagger}$, plus the premium received from selling the call option $\mathcal{P}(S(0),K)$. Both terms are __realized gains__ since we __must__ sell the shares at price $K$.
> * __Regime 2:__ In this case, $S(T) \leq K$, the stock price at expiration is below or equal to the strike price of the call option. The call option expires worthless, and we retain ownership of the shares. The profit consists of the (unrealized) difference between the stock price at expiration $S(T)$ and the share acquisition price $S^{\dagger}$, plus the premium received from selling the call option $\mathcal{P}(S(0),K)$.

## Cash Secured Puts
A cash secured put is an options strategy where an investor sells put options on an underlying asset while simultaneously setting aside enough cash to purchase the underlying asset if the option is exercised. This strategy is typically used when the investor is willing to buy the underlying asset at a lower price and wants to generate income from the option premiums.

Starting with the general portfolio expression for $|\mathcal{P}|$ legs of American-style __short put__ contracts on an underlying asset, all contracts have the same expiration date $T$ (which we assume is short enough to ignore time-value of money issues). The portfolio value consists of three terms: a cash position, a (potential) stock position, and an options position.
* __Cash Position:__ The cash position is simply the amount of cash held in the portfolio, denoted as $N_C$. For a cash secured put, we typically have $N_{C} \geq m \sum_{i=1}^{|\mathcal{P}|} n_i K_i$ to ensure sufficient funds to purchase the underlying stock if the puts are exercised.
* __Stock Position:__ The stock position is the value of the shares of the underlying stock $i$ held in the portfolio, denoted as $N^{(i)}_S S^{(i)}(t)$, where $N^{(i)}_S$ is the number of shares held of stock $i$ and $S^{(i)}(t)$ is the current share price of stock $i$ at time $t$. For a cash secured put, we typically have $N^{(i)}_S = 0$ since we do not own the underlying stock initially.
* __Options Position:__ The options position is the value of the options contracts held in the portfolio, denoted as $m \sum_{i=1}^{|\mathcal{P}|} \theta_i n_i \mathcal{P}_i(S,t)$, where $m$ is the contract multiplier (shares per contract), $\theta_i$ indicates long ($1$) or short ($-1$) position, $n_i$ is the number of contracts, and $\mathcal{P}_i(S,t)$ is the price of the option leg $i$ at time $t$.

Thus, the general portfolio value at time $t$ is given by:
$$
\begin{align*}
\Pi(S^{(1)},S^{(2)},\ldots,S^{(|\mathcal{P}|)},t) &= \underbrace{N_C}_{\text{Cash}} + \underbrace{\sum_{i=1}^{|\mathcal{P}|} N^{(i)}_S S^{(i)}(t)}_{\text{Stock Position}} + \underbrace{m\sum_{i=1}^{|\mathcal{P}|} \theta_i n_i \mathcal{P}_i(S^{(i)}(t),K_{i})}_{\text{Options Position}}.
\end{align*}
$$
where we use $S^{(i)}(t)$ to denote the price of stock $i$ at time $t$, and $\mathcal{P}_i(S^{(i)}(t),K_{i})$ to denote the price of option leg $i$ at time $t$ with strike price $K_i$. We have not explicitly written the dependence of the option price on time $t$ for clarity.

### Single cash secured put
The general expression is handy if we want to analyze a portfolio with multiple cash secured put legs, but let's simplify to a single cash secured put position to better understand the strategy. Consider a portfolio with one cash secured put position on stock `XYZ`. Further suppose we have sold $n$ put option contracts with strike price $K$ and expiration date $T$. 

At time $t=0$, we receive the option premium $m n \mathcal{P}(S(0),K)$ which is added to our cash position. At expiration $t=T$, the put option is either exercised (if $S(T) < K$) or expires worthless (if $S(T) \geq K$). 

* __Expires worthless:__ When the put expires worthless ($S(T) \geq K$), we have no obligation to purchase shares. Our final portfolio value is simply $\Pi(S,T) = N_C^{(0)} + mn\mathcal{P}(S(0),K)$, and our profit is $\hat{P}(S,T) = mn\mathcal{P}(S(0),K)$ since we keep the premium.
* __Exercised:__ When the put is exercised ($S(T) < K$), we must buy $mn$ shares at the strike price $K$. This results in a cash outflow of $mnK$. The shares we acquire are worth $mnS(T)$. Our final portfolio value becomes $\Pi(S,T) = (N_C^{(0)} + mn\mathcal{P}(S(0),K) - mnK) + mnS(T)$, representing our remaining cash plus the value of the stock we now own. The profit is $\hat{P}(S,T) = mn\mathcal{P}(S(0),K) - mnK + mnS(T) = mn[S(T) - K + \mathcal{P}(S(0),K)]$.

Combining both scenarios, the profit simplifies to:
$$
\boxed{
\begin{align*}
\hat{P}(S,T) &= \begin{cases}
m n\;\overbrace{\mathcal{P}(S(0),K)}^{\text{realized gain}}, & \text{if } S(T) \geq K\;\text{(put expires worthless)}\\
m n\;\underbrace{[S(T) - K + \mathcal{P}(S(0),K)]}_{\text{net P/L}}, & \text{if } S(T) < K\;\text{(put exercised)}
\end{cases}
\end{align*}}
$$

This piecewise function shows two interesting regimes:

> * __Regime 1:__ In this case, $S(T) \geq K$, the stock price at expiration is above or equal to the strike price of the put option. The put option expires worthless, and we retain the premium received from selling the put option $\mathcal{P}(S(0),K)$. This is a __realized gain__ since we keep the premium without any obligation to buy the stock.
> * __Regime 2:__ In this case, $S(T) < K$, the stock price at expiration is below the strike price of the put option. The put option is exercised, and we must buy the shares at the strike price $K$. The profit consists of the premium received from selling the put option $\mathcal{P}(S(0),K)$ minus the difference between the strike price $K$ and the stock price at expiration $S(T)$. This results in a __realized loss__ since we are obligated to buy the stock at a higher price than its market value.

## Delta of Covered Calls and Cash Secured Puts
Finally, let's briefly discuss the spot Delta ($\Delta$) of covered call and cash secured put portfolios. 
Suppose we want to compute the portfolio Delta at time $t$ with respect to leg $j$ of the portfolio. The portfolio Delta is given by:

$$
\begin{align*}
\Delta_{\text{net}}(S,t)&=\frac{\partial \Pi}{\partial S_{j}} \\
&= \frac{\partial N_{C}}{\partial S_{j}} + \sum_{i=1}^{|\mathcal{C}|} N^{(i)}_S \underbrace{\frac{\partial S^{(i)}(t)}{\partial S_{j}}}_{= ??} + m\sum_{i=1}^{|\mathcal{C}|} \theta_i n_i \frac{\partial \mathcal{P}_i(S^{(i)}(t),K_{i})}{\partial S_{j}} \\
&= 0 + \sum_{i=1}^{|\mathcal{C}|} N^{(i)}_S \delta_{ij} + m\sum_{i=1}^{|\mathcal{C}|} \theta_i n_i \Delta_i(S^{(i)}(t),K_{i})\delta_{ij} \\
\end{align*}
$$
where $\delta_{ij}$ is the Kronecker delta function, which is $1$ if $i=j$ and $0$ otherwise. 

> __Note on Delta Interpretation:__
>
> Portfolio Delta $\partial \Pi/\partial S_j$ measures the __instantaneous sensitivity__ of portfolio value to changes in asset $j$'s price, holding other prices fixed. This quantifies __direct exposure__ to each asset and is the standard metric for position sizing and hedging.
>
> However, when asset price changes are correlated (e.g., stocks in the same sector), total portfolio risk requires accounting for __combined effects__. Use portfolio variance $\mathbf{\Delta}^T \mathbf{\Sigma} \mathbf{\Delta}$ (where $\mathbf{\Sigma}$ is the covariance matrix of price changes) or risk measures like VaR/Expected Shortfall.

Thus, the portfolio Delta for movement of the price of underlying asset $j$ simplifies to:
$$
\Delta_{\text{net}}(S^{(j)},t) = N^{(j)}_S + m \theta_j n_j \Delta_j(S^{(j)}(t),K_{j}).
$$
where $\Delta_j(S^{(j)}(t),K_{j})$ is the Delta of option leg $j$ at time $t$. This is the same expression we derived previously for composite options contracts.

### Covered Call Delta
For a covered call position on a single underlying asset, we have $\theta_j = -1$ (short call), and $n_{j} = 1$ so the portfolio Delta becomes:
$$
\Delta_{\text{net}}(S,t) = N^{(j)}_S - m \Delta^{(c)}_j(S^{(j)}(t),K_{j}).
$$

> __When will a covered call be Delta neutral?__ 
> 
> A covered call position achieves Delta neutrality when $\Delta_{\text{net}}(S,t) = 0$, which requires:
> $$N^{(j)}_S = m \Delta^{(c)}_j(S^{(j)}(t),K_{j})$$
> 
> For a standard covered call with $N^{(j)}_S = m$ shares (one contract's worth), this occurs when the call option Delta equals 1, i.e., $\Delta^{(c)}_j \approx 1$. This happens when the call is __deep in the money__ ($S^{(j)}(t) \gg K_{j}$), where the option behaves almost identically to the underlying stock. In this scenario, the long stock position (+$m$ Delta) is offset by the short call position (−$m$ Delta), resulting in near-zero net directional exposure.

Let's check this out in the market. Can we see this behavior?

### Cash Secured Put Delta
A cash secured put position on a single underlying asset has $\theta_j = -1$ (short put), and $n_{j} = 1$ and behaves like a short put position (cash does not contribute to Delta), so the portfolio Delta becomes:
$$
\Delta_{\text{net}}(S,t) = - m \Delta^{(p)}_j(S^{(j)}(t),K_{j}).
$$

> __When will the cash secured put be Delta neutral?__ 
> 
> A cash secured put position achieves Delta neutrality when $\Delta_{\text{net}}(S,t) = 0$, which requires:
> $$\Delta^{(p)}_j(S^{(j)}(t),K_{j}) = 0$$
> 
> Since put option Deltas are negative and approach zero as they move out of the money, this condition is satisfied when the short put contract is __deep out of the money__, i.e., when $S^{(j)}(t) \gg K_{j}$, and thus $\Delta^{(p)}_j \approx 0$. In this scenario, the put option has minimal sensitivity to the underlying stock price movements. The position effectively has no directional exposure because the probability of exercise is negligible, and the option's value changes very little with small stock price movements.

## Summary
This lecture introduces covered calls and cash secured puts as strategies for generating income from options positions.

> __Key Takeaways:__
> 
> * __Covered calls combine stock ownership with short call options:__ The portfolio value includes cash, stock, and short call positions. At expiration, the profit depends on whether the stock price exceeds the strike price. If exercised, shares are sold at the strike price resulting in a realized gain. If the call expires worthless, the investor retains the shares and the premium as unrealized gain.
> * __Cash secured puts generate income without initial stock ownership:__ The portfolio value includes cash and short put positions. At expiration, if the stock price stays above the strike price, the put expires worthless and the premium is kept as a realized gain. If the stock price falls below the strike price, the investor must buy shares at the strike price resulting in a realized loss offset by the premium received.
> * __Portfolio Delta combines stock and options contributions:__ The net Delta equals the stock position plus the weighted option Delta for each leg. This calculation uses the Kronecker delta to isolate contributions from specific legs when portfolios contain multiple positions.

Understanding covered calls and cash secured puts enables construction of income-generating strategies using options positions combined with stock or cash.
