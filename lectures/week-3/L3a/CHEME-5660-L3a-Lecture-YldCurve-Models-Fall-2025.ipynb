## Term Structure of Interest Rates
The term structure of interest rates is a fundamental concept in finance that describes the relationship between the interest rates of Treasury securities and their maturities. 
> __Term structure of interest rates__ is the relationship between the _yields_ on fixed income debt securities of different maturities but similar credit quality, observed at a specific point in time. It typically illustrates how interest rates vary with the length of time until a bond's maturity.

In order to dig into the term structure of interest rates, we need to understand a few key concepts, let's start with the yield and the spot rate.

### Yields and Spot Rates for Zero-Coupon Instruments
The yield is the return on an investment, expressed as a percentage of the investment's cost. Suppose we are look at zero-coupon Treasury securities, which are bonds that do not pay periodic interest but are sold at a discount to their face value. 
> __Investment yield__: The investment yield on a zero-coupon bond is the difference between the purchase price and the face value, divided by the purchase price, multipled by a factor to annualize the yield.

> __Spot rate__: The spot rate is the discount rate used to calculate the present value of future cash flows from a zero-coupon bond. The relationship between the investment yield and the spot rate can be derived from the bond pricing formula.

The _annualized investment yield_ of a zero-coupon instrument with $t$ days to maturity (units: days), face (par) value $V_{P}$ (units: USD), and purchase price $V_{B}$ (units: USD) is given by:
$$
\begin{align*}
Y(t) &= \left(\frac{V_{P} - V_{B}}{V_{B}}\right) \times \left(\frac{365}{t}\right)\\
&= \left(\frac{V_{P}}{V_{B}} - 1\right) \times \left(\frac{365}{t}\right)\quad\Longrightarrow{T = \frac{t}{365}}\\
& = \underbrace{\left(\frac{V_{P}}{V_{B}} - 1\right)}_{\text{HPR}} \times \underbrace{\left(\frac{1}{T}\right)}_{\text{annualize}}\quad\blacksquare\\
\end{align*}
$$
where $\text{HPR}$ is the __holding period return__ and the $1/T$ term is an __annualization factor__. To better understand the relationship between the yield and the discount (spot) rate, we can rewrite the yield as:
$$
\begin{align*}
Y &= \left(\frac{V_{P}}{V_{B}} - 1\right) \times \left(\frac{1}{T}\right)\quad\Longrightarrow{V_{B}=\mathcal{D}_{nT,0}^{-1}(\bar{r})\;V_{P}}\\
Y & = \left(\frac{V_{P}}{\mathcal{D}_{nT,0}^{-1}(\bar{r})\;V_{P}} - 1\right)\times \left(\frac{1}{T}\right)\\
Y & = \left(\frac{1}{\mathcal{D}_{nT,0}^{-1}(\bar{r})} - 1\right)\times \left(\frac{1}{T}\right)\\
Y & = \left(\mathcal{D}_{nT,0}(\bar{r}) - 1\right)\times \left(\frac{1}{T}\right)\quad\Longrightarrow\text{solve for discount factor}\\
\mathcal{D}_{nT,0}(\bar{r}) &= 1 + Y\;T\quad\Longrightarrow\mathcal{D}_{nT,0}(\bar{r})=\left(1+\bar{r}/n\right)^{nT}\\
\left(1+\bar{r}/n\right)^{nT} & = 1 + Y\;T\quad\Longrightarrow\text{solve for}\;\bar{r}\\
\bar{r} & = n\left[\left(1 + Y\;T\right)^{1/(nT)} - 1\right]\quad\blacksquare
\end{align*}
$$
This derivation shows the relationship between the **investment yield** $Y$ (a simple annualized return calculation), 
the time to maturity $T$ (in years), the compounding frequency $n$ (number of compounding periods per year),
and the **spot rate** $\bar{r}$ (the discount rate used for discounting a zero-coupon pricing at auction).

> __Example__: For a 52-week zero-coupon Treasury Bill with semiannual compounding ($n=2$, $T=1$), the relationship between the discount rate and the investment yield can be expressed as:
>$$
>\boxed{
>\bar{r} = 2\left[\left(1 + Y\right)^{1/2} - 1\right] \quad\text{for}\;n=2,\;T=1\quad\text{(52-week zero-coupon)}
>}
>$$
>where $Y$ is the annualized investment yield and $\bar{r}$ is the corresponding __spot rate__ used in zero-coupon pricing calculations.

### Yield to Maturity (YTM) for Coupon-Bearing Notes (Bonds)
The yield to maturity (YTM) is a key concept for coupon-bearing notes (bonds). 
> __Yield to Maturity (YTM)__ represents the total return an investor can expect to earn if the note (bond) is held until maturity, taking into account both the bond's current market price and its future cash flows (coupon payments and face value at maturity).

The YTM for a coupon-bearing note (bond) is calculated using by computing the net present value (NPV) of the bond's __remaining__ cash flows, which include periodic coupon payments and the face value at maturity. The YTM is the _discount rate_ that makes the NPV of these cash flows equal to zero (where we include the purchase price of the bond as a cash flow event at time $t=0$). 
$$
\underbrace{\texttt{NPV}(T_{\text{rem}}, \bar{Y}, \bar{c})}_{=\,0} = \mathcal{D}^{-1}_{nT_{\text{rem}},0}(\bar{Y})\cdot{V_{P}}+C\cdot\sum_{j=1}^{n{T_{\text{rem}}}}\mathcal{D}^{-1}_{j,0}(\bar{Y}) - V_{B} = 0
$$
where $C$ is the coupon payment, $V_{P}$ is the face value of the bond, and $V_{B}$ is the current market price of the bond, $T_{\text{rem}}$ is the remaining time to maturity (in years), and $\bar{Y}$ is the yield to maturity. 

> __Auction__: At auction, the YTM  is equal to the discount rate used to compute the present value of the bond's cash flows, which include the periodic coupon payments and the face value at maturity. This is because $T = T_{\text{rem}}$ at the time of auction, and the bond is priced to yield the YTM at that time. 
>
>__Secondary Market__: In the secondary market, the YTM is the discount rate that equates the present value of the bond's future cash flows to its current market price, given the remaining time to maturity $T_{\text{rem}}$ and the bond's coupon payments. The YTM reflects the bond's current market conditions, including interest rates and credit risk.


The YTM can be computed using various numerical methods, however, let's look at a particualy simple method, the __Secant Method__.

#### Estimation of YTM using the Secant Method
The secant method is a numerical technique used to find the root of a function of single variable, which in this case is the YTM. Consider the general case of finding $x\in\mathbb{R}$ such that $f(x) = 0$.  

Let's look at some pseudo-code to illustrate the secant method for estimating the YTM.

__Initialization:__ Given two initial guesses $x_0$ and $x_1$, a tolerance level $\epsilon>0$, a maximum number of iterations $\texttt{maxiter}$, a convergence flag $\texttt{converged}\gets\texttt{false}$ and an iteration counter $i\gets{1}$.

> __Parameter Selection Rules of Thumb:__
>
> **Tolerance ($\epsilon$)**: For financial calculations like YTM, a tolerance of $\epsilon = 10^{-6}$ to $10^{-8}$ is typically appropriate. This provides accuracy to about 6-8 decimal places, which is sufficient for most practical applications. For YTM calculations, $\epsilon = 10^{-6}$ corresponds to about 0.0001% accuracy in the yield estimate.
>
>**Maximum iterations ($\texttt{maxiter}$)**: For well-behaved functions like the NPV equation for YTM, the secant method typically converges within 10-20 iterations. A safe choice is $\texttt{maxiter} = 50$ to $100$. If convergence hasn't occurred by then, it usually indicates a problem with the initial guesses or the function itself.
>
> **Initial guesses**: For YTM calculations, reasonable starting values are often $x_0 = 0.01$ (1%) and $x_1 = 0.10$ (10%), which bracket typical bond yields. The secant method is generally robust to the choice of initial guesses as long as they're not too close together.

While not $\texttt{converged}$ and $i\leq\texttt{maxiter}$ __do__:
1. Compute the next estimate $x_{i+1}$ using the secant formula:
   $$
   x_{i+1} = x_i - f(x_i) \cdot \frac{x_i - x_{i-1}}{f(x_i) - f(x_{i-1})}
   $$
   where $f(x_i)$ and $f(x_{i-1})$ are the function values at the current and previous estimates, respectively.
2. Check for convergence: if $|x_{i+1} - x_i| < \epsilon$, set $\texttt{converged}\gets\texttt{true}$. Otherwise, continue iterating.
3. Update for next iteration: set $x_{i-1} \gets x_i$, $x_i \gets x_{i+1}$, and increment $i \gets i + 1$.

__Convergence Properties for YTM Calculations:__

The secant method exhibits **superlinear convergence** for YTM calculations under typical conditions, with several important characteristics:

**Convergence Rate**: The secant method has a convergence order of approximately 1.618 (the golden ratio), which means it converges faster than linear methods but slower than Newton's method (quadratic convergence). In practice, this means the number of accurate decimal places increases by about 60% with each iteration once near the solution.

**Why is Secant well-suited for YTM calculations**: The NPV function for coupon note (bond) pricing is smooth, continuous, and monotonically decreasing with respect to the discount rate. This makes it an ideal candidate for the secant method because:
  - The function has a unique root (the YTM)
  - There are no discontinuities or sharp turns that could cause convergence issues
  - The derivative doesn't change sign, ensuring stable convergence

**Typical Convergence Behavior**: For YTM calculations, the secant method typically converges within a _few_ iterations, i.e., not thousands when starting with reasonable initial guesses (e.g., 1% and 10%). 

While Newton's method converges faster, it requires computing the derivative of the NPV function. The secant method achieves good convergence _without_ this additional computational complexity, making it attractive for YTM calculations.

## Yield Curve

### Par Yield Curve Rates

The _par yield_ for a given maturity is the coupon rate that makes a newly issued (standard) coupon-bearing bond trade exactly at par (face) value when priced with the current set of discount factors (spot rates). The par yield curve is the function that maps maturity $T$ to the par coupon rate $\bar c_{\text{par}}(T)$.

Intuitively: a par bond is priced so that its present value equals its face value. For a given set of spot (discount) factors, we can solve for the coupon rate that achieves this equality — that coupon rate is the par yield.

Let $V_{P}$ denote the face (par) value (normalize $V_{P}=1$ for simplicity), let $n$ be the number of coupon payments per year, and let $\mathcal{D}^{-1}_{j,0}(\bar r)$ denote the discrete-period discount factor for period $j$ (as used earlier, e.g. $\mathcal{D}^{-1}_{j,0}(\bar r)=(1+\bar r/n)^{-j}$ when the spot rate is flat for that maturity). For a bond with maturity $T$ (years) there are $nT$ coupon payments (i.e., $n\cdot T$ payments).

The price of a coupon-bearing bond with coupon rate $\bar c$ (annualized) is

$$
V_B(\bar c,T) 
= C\sum_{j=1}^{nT}\mathcal{D}^{-1}_{j,0}(\bar r) + V_{P}\,\mathcal{D}^{-1}_{nT,0}(\bar r),
$$

where the coupon payment per period is $C=(\bar c/n)\,V_{P}$. The _par yield_ $\bar c_{\text{par}}(T)$ satisfies $V_B(\bar c_{\text{par}},T)=V_{P}$. Setting $V_{P}=1$ and solving for $\bar c_{\text{par}}(T)$ gives:

$$
1 = \frac{\bar c_{\text{par}}(T)}{n}\sum_{j=1}^{nT}\mathcal{D}^{-1}_{j,0}(\bar r) + \mathcal{D}^{-1}_{nT,0}(\bar r)
$$

and therefore

$$
\boxed{\displaystyle \bar c_{\text{par}}(T) 
= n\cdot\frac{1 - \mathcal{D}^{-1}_{nT,0}(\bar r)}{\sum_{j=1}^{nT}\mathcal{D}^{-1}_{j,0}(\bar r)}.}
$$

This formula is the mathematical definition of the par yield: the annual coupon rate which, when paid each period and discounted by the current discount factors, yields exactly the par value.

A few important remarks:

* __Normalization__: If $V_{P}\neq 1$ then the same algebra applies and $\bar c_{\text{par}}(T)$ is independent of the nominal par amount because both sides scale with $V_{P}$.
* __Compounding / payment frequency__: The factor $n$ converts the per-period coupon into an annualized coupon rate; the formula above reduces to familiar annual-coupon expressions for $n=1$.
* __Interpretation__: The numerator $1-\mathcal{D}^{-1}_{nT,0}$ measures the fraction of value that is _not_ paid by the final principal payment (i.e., the portion that must be supplied by coupons), while the denominator sums the present value of the periodic coupon annuity. Their ratio (times $n$) is the coupon rate needed to make the coupon annuity plus discounted principal equal to par.

## STRIPs (Separate Trading of Registered Interest and Principal Securities)

STRIPs are the market instruments that correspond to individual scheduled cash flows of coupon-bearing Treasury securities. When dealers or the Treasury "strip" a coupon bond, each coupon and the final principal payment can be traded separately as a zero‑coupon claim. STRIPs therefore offer direct market observations of the zero (spot) prices (discount factors) for specific payment dates.

**Notation (consistent with earlier sections):** earlier we used the discrete-period discount factor notation $\mathcal{D}^{-1}_{j,0}$ for the present price of a unit payment at period $j$. In this subsection we define the STRIP price

$$
P_j \triangleq \mathcal{D}^{-1}_{j,0},
$$

so $P_j$ and $\mathcal{D}^{-1}_{j,0}$ are identical objects (the market price today of one unit paid at period $j$). This keeps the STRIPs notation consistent with the par-yield section where discount factors appeared as $\mathcal{D}^{-1}_{j,0}$.

A reminder on indexing: if a bond has maturity $T$ years and $n$ coupon payments per year, then the final payment occurs at period $m=nT$. We therefore write sums to $nT$ (or equivalently to $m$) and will sometimes use $m$ as a shorthand when convenient.

Because STRIPs are zeros, the relation between par yields and STRIP prices is immediate. Using the par‑bond pricing identity with $V_{P}=1$ (and replacing discount factors by $P_j$):

$$
1 = \frac{\bar c_{\text{par}}(T)}{n}\sum_{j=1}^{nT}P_j + P_{nT}.
$$

If we observe the vector of STRIP prices $\{P_j\}_{j=1}^{nT}$ directly, the par yield $\bar c_{\text{par}}(T)$ is computed by the same formula for par yields by replacing discount factors with $P_j$.

Conversely, if we observe market par yields for standard coupon maturities (as is typical), we can recover the STRIP prices (the zero curve) by _bootstrapping_. The elementary bootstrapping recursion for period $m$ (where $m=nT$) is obtained by isolating $P_m$ from the par identity for the bond maturing at period $m$:

$$
1 = \frac{\bar c_{\text{par}}(T_m)}{n}\sum_{j=1}^{m-1}P_j + \Big(\frac{\bar c_{\text{par}}(T_m)}{n} + 1\Big)P_m.
$$

Solving for $P_m$ gives the standard bootstrap step:

$$
\boxed{\displaystyle P_m 
= \frac{1-\dfrac{\bar c_{\text{par}}(T_m)}{n}\sum_{j=1}^{m-1}P_j}{1+\dfrac{\bar c_{\text{par}}(T_m)}{n}}.}
$$

Start with the shortest maturity (smallest $m$) where the sum over earlier $P_j$ is empty and solve for $P_1$. Proceed recursively: each observed market par yield yields one equation, which after substitution of already-determined earlier $P_j$ gives the next $P_m$.

A few teaching notes you may want to emphasize to students:

* STRIPs provide the cleanest, arbitrage‑free definition of the spot curve because they are pure zeros. Market STRIP prices are therefore the most direct input to exact valuation and risk metrics.
* Par yields are what market participants typically quote for coupon issues; bootstrapping is the practical conversion from these convenient market quotes to the STRIP/spot curve used for valuation.
* If observed STRIP prices (when available) and par‑implied STRIP prices disagree, arbitrage opportunities arise and market makers will act to remove the discrepancy.
