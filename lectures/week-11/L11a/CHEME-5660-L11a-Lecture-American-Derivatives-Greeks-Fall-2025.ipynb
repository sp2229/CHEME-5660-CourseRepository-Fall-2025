## Concept Review: Intrinsic and Extrinsic Value of Options
In this last lecture, we did a deep dive into American options premium and its components: intrinsic value and extrinsic value.

* __Intrinsic Value__: The intrinsic value of an American option is the __immediate exercise value__. For a call option, it's calculated as $\max(0, S - K)$, where $S$ is the current stock price and $K$ is the strike price. For a put option, it's $\max(0, K - S)$.
* __Extrinsic Value__: The extrinsic value, also known as time value, represents the additional value of the option based on the potential for future price movements before expiration. It is calculated as the difference between the total option price and its intrinsic value.

The extrinsic value $E_{V}(t)$ can be expressed mathematically as:
$$
\boxed{
\begin{align*}
E_{V}(t) = \text{Option Price} - \text{Intrinsic Value}\quad\blacksquare
\end{align*}}
$$
or in expanded form:
$$
\boxed{
\begin{align*}
E_{V}(t) &= \begin{cases}
\mathcal{P}_{c}(K,S(0)) - \max(0, S(t) - K) & \text{(for Call Options)} \\
\mathcal{P}_{p}(K,S(0)) - \max(0, K - S(t)) & \text{(for Put Options)}
\end{cases}
\end{align*}}
$$
This creates an interesting edge case: when the option is __out of the money__, i.e., the option has no intrinsic value, the entire option price is extrinsic value. Why would we be willing to pay for an option that has no intrinsic value?

## Options Premiums and Market Conditions
The central question we will address in this lecture is: "How do American options premiums change with market conditions?" To answer this, we will explore the sensitivity of American options premiums to various factors, including: underlying asset price, volatility, time to expiration, interest rates (risk-free rate), and dividends.

This question is answered (at least locally) by computing sensitivity coefficients of the options premium collectively known as the __Greeks.__

> __The Greeks__ 
> 
> The Greeks are a set of (local) measures that describe how the price of an option changes in response to changes in various parameters. They are __essential__ tools for options traders to assess and manage the risks associated with options positions.
> - **Delta ($\Delta$)**: Measures the sensitivity of the option's price to `+1 USD/share` change in the underlying asset's price. However, $\Delta$ also represents the hedge ratio, i.e., the number of shares of the underlying asset needed to hedge one option contract, and the probability that the option will finish in-the-money at expiration. Wow!
> - **Theta ($\Theta$)**: Measures the sensitivity of the option's price to  `1-day decrease` in the lifetime of the contract (time decay). Theta benefits option sellers, as options lose value as expiration approaches, all else being equal.
> - **Vega ($\nu$)**: Measures the sensitivity of the option's price to a `+1%` change in the volatility of the underlying asset. The higher the volatility, the higher the option premium, as there is a greater chance of significant price movements.
> - **Gamma ($\Gamma$)**: Measures the sensitivity of $\Delta$ to changes in a `+1 USD/share` change in the underlying asset's price. This is the second derivative of the option price with respect to the underlying asset price.
> - **Rho ($\rho$)**: Measures the sensitivity of the option's price to a `+1%` change in the risk-free interest rate. This is generally a minor effect for most options, but can be more significant for longer-term options.
>
> When trading, most trading platforms will compute these for you automatically!

### Derivation
Suppose we have an American option contract of type $\tau = \left\{c,p\right\}$, written against an underlying asset with share price $S$, that has $t$ days to expiration, and (implied) volatility $\sigma$. We can take the total differential of the premium of that contract, $\mathcal{P}_{\tau} = \mathcal{P}_{\tau}(S, \sigma, t, r)$, where $r$ is the risk-free interest rate, which gives us:
$$
\begin{align*}
d\mathcal{P}_{\tau} &= \underbrace{\left(\frac{\partial \mathcal{P}_{\tau}}{\partial S}\right)}_{=\Delta}\;dS + 
\underbrace{\left(\frac{\partial \mathcal{P}_{\tau}}{\partial \sigma}\right)}_{=\nu}\;d\sigma + 
\underbrace{\left(\frac{\partial \mathcal{P}_{\tau}}{\partial t}\right)}_{=\Theta}\;dt + \underbrace{\left(\frac{\partial \mathcal{P}_{\tau}}{\partial r}\right)}_{=\rho}\;dr + \frac{1}{2}\underbrace{\left(\frac{\partial^{2} \mathcal{P}_{\tau}}{\partial S^{2}}\right)}_{=\Gamma}\;dS^{2}\\
& = \Delta_{\tau}\;dS + \nu_{\tau}\;d\sigma + \Theta_{\tau}\;dt + \rho_{\tau}\;dr + \frac{1}{2}\;\Gamma_{\tau}\;dS^{2}
\end{align*}
$$
where all the partial derivatives are evaluated at the __current market conditions__, the current values of the parameters: $S$, $\sigma$, $t$, and $r$, i.e., at the current market conditions. Let's walk through the __major Greeks__ and see how they affect the American options premium: $\Delta$, $\Theta$, and $\nu$.

### Theta $\Theta$: Time decay and its impact on options premium.
The $\Theta$ of an option measures the sensitivity of the option's price to the passage of time. It represents the rate at which the option's premium decreases as the expiration date approaches, all else being equal. This phenomenon is known as time decay.

Time decay is particularly important for American options, as they can be exercised at any time before expiration. As time passes, the likelihood of the option finishing in-the-money changes, and this is reflected in the option's premium. Generally, the closer an option is to expiration, the faster its time decay.

### Vega $\nu$: Understanding Volatility's role in options pricing.
The $\nu$ of an option measures the sensitivity of the option's price to changes in the implied volatility of the underlying asset. What is the implied volatility? 

> __Implied Volatility (IV)__
>
> Implied volatility is the market's expectation of the future annualized volatility of the underlying asset. It is derived from the market price of the option itself, using an options pricing model (e.g., CRR). Higher IV generally leads to higher option premiums, as it increases the potential for larger price movements in the underlying asset. In general, the IV is larger than the historical volatility (HV) of the underlying asset, as it includes a risk premium for uncertainty about future price movements.

We can use the IV to get a __back of the envelope estimate__ of the expected price range of the underlying asset over a given time horizon. For example, suppose the current price of the underlying asset is $S_{0}$, and the implied volatility is $\sigma_{IV}$ (written as a decimal). Over a time horizon of $T$-days, we can estimate the expected price range of the underlying asset as:
$$
\boxed{
\begin{align*}
S_{T} & \approx S_{0}\left(1 \pm z_{\alpha}\left(\sqrt{\frac{T}{252}}\right)\sigma_{IV}\right)\\
\end{align*}}
$$
where $z_{\alpha}$ is the z-score corresponding to the desired confidence level (e.g., for a 95% confidence level, $z_{\alpha} \approx 1.96$), and 252 is the typical number of trading days in a year. Let's check out a quick example of the IV using our NVDA options data.

## The Many Secret Lives of Delta $\Delta$.
The $\Delta$ of an option measures how much the price of the option changes for a `+1 USD/share` change in the price of the underlying asset. However, it has two other important interpretations:

> __Delta Interpretations:__
> 
> * __Hedging Ratio:__ Delta can be interpreted as the number of shares of the underlying asset needed to hedge one option contract. For example, if an option contract has $\pm\Delta$, it means that for every $n$ contracts held, an investor would need to hold $\mp{100}\cdot{n_{i}}\cdot\Delta$ shares of the underlying asset to be delta-neutral. A delta-neutral position is one where the overall portfolio value is insensitive to small changes in the price of the underlying asset.
> * __Probability Measure:__ Delta can also be interpreted as the probability that the option will expire in-the-money. For example, a delta of 0.7 for a call option suggests a 70% chance that the option will be exercised profitably at expiration.

We've already seen an example of the hedge ratio interpretation of $\Delta$ in the last example notebook. Let's look the probability interpretation, in particular, we can use $\Delta$ to estimate the probability of profit (POP) for an option contract:

> __Probability of Profit (POP):__ 
>
> The probability of profit (POP) is the probability that at the __expiration of the (composite or single) contract__ at time $t = T$, the share price of the underlying asset $S_{T}$ is at least `0.01` USD/share greater than the breakeven points of the (composite or single) contract, i.e.,
> $$
> \boxed{
> \begin{align*}
> \text{POP} & = \mathbb{P}\left(\hat{P}_{\mathcal{C}} > 0\right)
> \end{align*}}
> $$
> where $\hat{P}_{\mathcal{C}}$ is the profit of the (single or composite) contract at expiration. Let's check at the conjecture that $\Delta \approx \text{POP}$ using an example.
