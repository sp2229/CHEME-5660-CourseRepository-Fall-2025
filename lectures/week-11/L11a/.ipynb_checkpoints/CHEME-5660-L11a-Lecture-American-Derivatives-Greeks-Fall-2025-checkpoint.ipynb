## Concept Review: Intrinsic and Extrinsic Value of Options
In the last lecture, we did a deep dive into American options premium and its components: intrinsic value and extrinsic value.

* __Intrinsic Value__: The intrinsic value of an American option is the immediate exercise value. For a call option, it's calculated as $\max(0, S_{t} - K)$, where $S_{t}$ is the current stock price and $K$ is the strike price. For a put option, it's $\max(0, K - S_{t})$.
* __Extrinsic Value__: The extrinsic value, also known as time value, represents the additional value of the option based on the potential for future price movements before expiration. It is calculated as the difference between the total option price and its intrinsic value.

The extrinsic value $E_{V}(t)$ can be expressed mathematically as:
$$
\boxed{
\begin{align*}
E_{V}(t) = \text{Option Price} - \text{Intrinsic Value}\quad\blacksquare
\end{align*}}
$$
or in expanded form:
$$
\boxed{
\begin{align*}
E_{V}(t) &= \begin{cases}
\mathcal{P}_{c} - \max(0, S_{t} - K) & \text{(for Call Options)} \\
\mathcal{P}_{p} - \max(0, K - S_{t}) & \text{(for Put Options)}
\end{cases}
\end{align*}}
$$
This creates an interesting edge case: when the option is __out of the money__, i.e., the option has no intrinsic value, the entire option price is extrinsic value. Why would we be willing to pay for an option that has no intrinsic value?

## Options Premiums and Market Conditions
The central question we will address in this lecture is: "How do American options premiums change with market conditions?" To answer this, we will explore the sensitivity of American options premiums to various factors, including underlying asset price, volatility, time to expiration, interest rates (risk-free rate), and dividends.

This question is answered (at least locally) by computing sensitivity coefficients of the options premium collectively known as the __Greeks.__

> __The Greeks__ 
> 
> The Greeks are a set of (local) measures that describe how the price of an option changes in response to changes in various parameters. They are essential tools for options traders to assess and manage the risks associated with options positions.
> - **Delta ($\Delta$)**: Measures the sensitivity of the option's price to a `+1 USD/share` change in the underlying asset's price. However, $\Delta$ also represents the hedge ratio, i.e., the number of shares of the underlying asset needed to hedge one option contract, and the probability that the option will finish in-the-money at expiration.
> - **Theta ($\Theta$)**: Measures the sensitivity of the option's price to a `1-day decrease` in the lifetime of the contract (time decay). Theta benefits option sellers, as options lose value as expiration approaches, all else being equal.
> - **Vega ($\nu$)**: Measures the sensitivity of the option's price to a `+1%` change in the volatility of the underlying asset. The higher the volatility, the higher the option premium, as there is a greater chance of significant price movements.
> - **Gamma ($\Gamma$)**: Measures the sensitivity of $\Delta$ to a `+1 USD/share` change in the underlying asset's price. This is the second derivative of the option price with respect to the underlying asset price.
> - **Rho ($\rho$)**: Measures the sensitivity of the option's price to a `+1%` change in the risk-free interest rate. This is generally a minor effect for most options, but can be more significant for longer-term options.
>
> When trading, most trading platforms will compute these for you automatically!

### Derivation
Suppose we have an American option contract of type $\tau = \left\{c,p\right\}$, written against an underlying asset with share price $S$, that has $t$ days to expiration, and (implied) volatility $\sigma$. We can take the total differential of the premium of that contract, $\mathcal{P}_{\tau} = \mathcal{P}_{\tau}(S, \sigma, t, r)$, where $r$ is the risk-free interest rate, which gives us:
$$
\begin{align*}
d\mathcal{P}_{\tau} &= \underbrace{\left(\frac{\partial \mathcal{P}_{\tau}}{\partial S}\right)}_{=\Delta}\;dS + 
\underbrace{\left(\frac{\partial \mathcal{P}_{\tau}}{\partial \sigma}\right)}_{=\nu}\;d\sigma + 
\underbrace{\left(\frac{\partial \mathcal{P}_{\tau}}{\partial t}\right)}_{=\Theta}\;dt + \underbrace{\left(\frac{\partial \mathcal{P}_{\tau}}{\partial r}\right)}_{=\rho}\;dr + \frac{1}{2}\underbrace{\left(\frac{\partial^{2} \mathcal{P}_{\tau}}{\partial S^{2}}\right)}_{=\Gamma}\;dS^{2}\\
& = \Delta_{\tau}\;dS + \nu_{\tau}\;d\sigma + \Theta_{\tau}\;dt + \rho_{\tau}\;dr + \frac{1}{2}\;\Gamma_{\tau}\;dS^{2}
\end{align*}
$$
where all the partial derivatives are evaluated at the current market conditions, i.e., the current values of the parameters: $S$, $\sigma$, $t$, and $r$. Let's walk through the major Greeks and see how they affect the American options premium: $\Delta$, $\Theta$, and $\nu$.

## The Secret Lives of Delta $\Delta$
The $\Delta$ of an option measures how much the price of the option changes for a `+1 USD/share` change in the price of the underlying asset. However, it has two other important interpretations:

> __Delta Interpretations:__
> 
> * __Hedging Ratio:__ Delta can be interpreted as the number of shares of the underlying asset needed to hedge one option contract. For example, if an option contract has $\Delta$, it means that for every $n$ contracts held, an investor would need to hold ${100}\cdot{n}\cdot\Delta$ shares of the underlying asset to be delta-neutral. A delta-neutral position is one where the overall portfolio value is insensitive to small changes in the price of the underlying asset.
> * __Probability of Exercise:__ Delta can also be interpreted as the probability that the option will expire in-the-money (ITM). For example, a delta of 0.7 for a call option suggests a 70% chance that the share price at expiration will be at least one cent above the strike price.

We've already seen an example of the hedge ratio interpretation of $\Delta$ in the last example notebook. Let's look at the probability interpretation of $\Delta$. In particular, we can use $\Delta$ to estimate the probability of exercise: 

> __Probability of Exercise (POE):__
>
> The probability of exercise (POE) is the probability that at the __expiration of the (composite or single) contract__ at time $t = T$, the share price of the underlying asset $S_{T}$ is at least `0.01` USD/share greater than the strike price $K$ of the (composite or single) contract, i.e.,
> $$
\boxed{
\begin{align*}
\text{POE} & = \mathbb{P}\left(S_{T} > K + 0.01\right)
\end{align*}}
$$
> We'll check the conjecture that $\Delta \approx \text{POE}$ using an example.

We can also compute the __Probability of Profit (POP)__ for an option contract, which is closely related to the probability of exercise.

> __Probability of Profit (POP):__ 
>
> The probability of profit (POP) is the probability that at the expiration of the (composite or single) contract at time $t = T$, the profit of the contract is positive, i.e.,
> $$
\boxed{
\begin{align*}
\text{POP} & = \mathbb{P}\left(\hat{P}_{\mathcal{C}} > 0\right)
\end{align*}}
$$
> where $\hat{P}_{\mathcal{C}}$ is the profit of the (single or composite) contract at expiration.

## Summary
This lecture introduces the Greeks as sensitivity measures that quantify how American options premiums respond to changes in market conditions.

> __Key Takeaways:__
> 
> * __The Greeks quantify option price sensitivity:__ Delta measures sensitivity to underlying price changes, Theta captures time decay, Vega measures volatility sensitivity, Rho tracks interest rate sensitivity, and Gamma measures the rate of change in Delta. These measures are derived from the total differential of the options premium function evaluated at current market conditions.
> * __Delta serves three distinct purposes:__ Delta represents the rate of change in option price per dollar change in underlying price, determines the hedge ratio for constructing delta-neutral positions, and approximates the probability that the option will expire in-the-money.
> * __Probability of profit differs from probability of exercise:__ Probability of exercise estimates whether the option will finish in-the-money at expiration, while probability of profit accounts for the premium paid and estimates whether the position will generate a net profit at expiration.

Understanding the Greeks and probability measures enables assessment of option positions and their exposure to different market risk factors.
